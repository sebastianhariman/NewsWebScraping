### **Web Scraping**

The next step is to scrape information from the news page. In this step, Beautiful Soup library is used to scrape information from each media

```python
# Kompas
for row in Kompas:
  r = requests.get(row[0])

  soup = BeautifulSoup(r.content, 'html5lib')

  article = soup.findAll('div', {'class': 'read__content'})

  paragraph = []
  for par in article:
    lines = par.findAll('p')
    for line in lines:
      # Exclude not needed elements
      if 'Baca juga' not in line.text and line.text.strip():
        a = line.find('a')
        if a:
          a.decompose()
        paragraph.append(line.text)

  row[0] = paragraph
  data.append(row)
```

```python
#CNN Indonesia
for row in CNN:
  r = requests.get(row[0])

  soup = BeautifulSoup(r.content, 'html5lib')

  article = soup.findAll('div', {'class': 'detail-text'})

  paragraph = []
  for par in article:
    lines = par.findAll('p')
    for line in lines:
      # Exclude not needed elements
      if 'ADVERTISEMENT' not in line.text and 'SCROLL TO CONTINUE WITH CONTENT' not in line.text and line.text.strip():
        paragraph.append(line.text)

  row[0] = paragraph
  data.append(row)
```

```python
#Detik
for row in Detik:
  r = requests.get(row[0])

  soup = BeautifulSoup(r.content, 'html5lib')

  article = soup.findAll('div', {'class': 'detail__body-text'})

  paragraph = []
  for par in article:
    lines = par.findAll('p')
    for line in lines:
      # Exclude not needed elements
      if 'ADVERTISEMENT' not in line.text and 'SCROLL TO CONTINUE WITH CONTENT' not in line.text and line.text.strip():
        paragraph.append(line.text)

  row[0] = paragraph
  data.append(row)
```

```python
#Liputan6
for row in Liputan6:
  r = requests.get(row[0])

  soup = BeautifulSoup(r.content, 'html5lib')

  article = soup.findAll('div', {'class': 'article-content-body__item-content'})

  paragraph = []
  for par in article:
    lines = par.findAll('p')
    for line in lines:
      # Exclude not needed elements
      if '* ' not in line.text and 'Baca Juga' not in line.text and 'Advertisement' not in line.text and line.text.strip():
        paragraph.append(line.text)

  row[0] = paragraph
  data.append(row)
```

## **Text Preprocessing**

### **Text Cleaning**

The first step of text preprocessing is text cleaning. In this process, the raw text will be cleaned. This process includes lowering text, removing punctuation, symbols, extra spaces, digits, urls, and unnecessary words.

```python
import re
```

```python
def clean_text(df, text_field):
  df[text_field] = df[text_field].str.lower()
  df[text_field] = df[text_field].apply(lambda elem: re.sub(r'xa0', ' ', elem))
  df[text_field] = df[text_field].apply(lambda elem: re.sub(r'(http[s]?\://\S+)|(#\S+)|(@\S+)|kompas\S+|liputan6\S+|baca juga\S+|\d+|(\n)+|(\[[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)', '', elem))
  df[text_field] = df[text_field].apply(lambda elem: re.sub(r'(\s){2,}', ' ', elem))
  df[text_field] = df[text_field].apply(lambda elem: re.sub(r'^(\s+)|(\s+)$', '', elem))

  return df
```

```python
clean_df = clean_text(raw_df, "Berita")
clean_df
```

### **Tokenization**

The next step is to tokenize the paragraph/text into smaller units, which is word. This process is done with the help of NLTK library.

```python
filtered_df = cleaned_df
```

```python
# Import libraries
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize

# Function to perform tokenization
def tokenize_text(text):
  tokens = nltk.tokenize.word_tokenize(text)
  return tokens
```

```python
filtered_df['Berita'] = filtered_df['Berita'].apply(tokenize_text)
```

### **Stemming**

It is a text normalization process to reduce words to their base form by removing suffixes or prefixes from words. This process is done with the help of Sastrawi library, which is a library that contains words in Bahasa Indonesia.

```python
!pip install Sastrawi
```

```python
# Import library
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

factory = StemmerFactory()
stemmer = factory.create_stemmer()
```

```python
# Function to perform token stemming
def stem_tokens(tokens):
    stemmed_tokens = [stemmer.stem(token) for token in tokens]
    return stemmed_tokens
```

```python
filtered_df['Berita'] = filtered_df['Berita'].apply(stem_tokens)
```

```python
filtered_df
```

### **Removing Stop Words**

This process aims to filter out stop words as they considered to be of little value in text analysis. This process is done with the help of NLTK library + additional custom Indonesia stop words found from github.

```python
# Custom Indonesian Stop Words

with open('id.stopwords.02.01.2016.txt') as f:
  custom_stop = f.read().splitlines()
```

```python
# Remove Stop Words
nltk.download('stopwords')
from nltk.corpus import stopwords

# Stop Words NLTK extended with custom stop words
stopwords = stopwords.words('indonesian')
stopwords.clear()
stopwords.extend(custom_stop)
stopwords
```

```python
# Function to perform stop words removal
def remove_stopwords(tokens):
    filtered_tokens = [word for word in tokens if word not in stopwords]
    return filtered_tokens
```

```python
filtered_df['Berita'] = filtered_df['Berita'].apply(remove_stopwords)
```

```python
# Import Libraries
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, classification_report, confusion_matrix

from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
```

## **Model and Evaluation**

The machine learning model is built using two algorithms: Support Vector Machine (SVM) and Random Forest. The two algorithm will be evaluated by comparing several aspects: precision, recall, f1-score, and support.

### **Text Representation**

Before the model is built, text representation is performed. It is a process of converting textual data into a numerical format that can be understood and processed by machine learning algorithms and models. The process is done using two methods: TfidVectorizer and CountVectorizer.

#### **TfidfVectorizer**

```python
# TfidfVectorizer

# Initialize a TfidfVectorizer object with vector size: 50 and min frequency: 3
tfidf = TfidfVectorizer(max_features=50, min_df=3)

tfidf_matrix = tfidf.fit_transform(data['Berita'])

tfidf_df = pd.DataFrame(data=tfidf_matrix.toarray(), columns=tfidf.get_feature_names_out())
tfidf_df.head()
```

```python
# Train test split

X_tfidf = tfidf_df
y_tfidf = data['Label']
X_train_tfidf, X_test_tfidf, y_train_tfidf, y_test_tfidf = train_test_split(X_tfidf, y_tfidf, test_size=0.2, random_state=36)
```

```python
print("Training Data:")
print(X_train_tfidf.head())
print(y_train_tfidf.head())

print("Test Data:")
print(X_test_tfidf.head())
print(y_test_tfidf.head())
```

```python
# Support Vector Machine
tfidf_svm = SVC()

tfidf_svm.fit(X_train_tfidf, y_train_tfidf)

tfidf_svm_predict = tfidf_svm.predict(X_test_tfidf)

tfidf_svm_accuracy = accuracy_score(y_test_tfidf, tfidf_svm_predict)

print(tfidf_svm_accuracy)
print(classification_report(y_test_tfidf, tfidf_svm_predict))
```

The classification report above shows that SVM give 100% in all metrics for predicting all labels. This could indicates that there might be an issue with the data or the way the model is evaluated as it is quite rare to get 100% accuracy. This problem could occur due to small number of data is used.

```python
# Random Forest
tfidf_rf = RandomForestClassifier()

tfidf_rf.fit(X_train_tfidf, y_train_tfidf)

tfidf_rf_predict = tfidf_rf.predict(X_test_tfidf)

tfidf_rf_accuracy = accuracy_score(y_test_tfidf, tfidf_rf_predict)

print(tfidf_rf_accuracy)
print(classification_report(y_test_tfidf, tfidf_rf_predict))
```

The classification report above shows that Random Forest give a high number on the 'Politics' metrics, medium number on the 'Entertainment' metrics, and small number on the 'Sports' metrics. And overall it gives 57% accuracy. This report could shows that the 'Entertainment' is underfitted. It could occur due to contents found in the 'Entertainment' is more variative so the model need to learn a bigger data.

#### **CountVectorizer**

```python
# CountVectorizer

# # Initialize a CountVectorizer object with vector size: 50 and min frequency: 3
cv = CountVectorizer(max_features=50, min_df=3)

cv_matrix = cv.fit_transform(data['Berita'])

cv_df = pd.DataFrame(data=cv_matrix.toarray(), columns=cv.get_feature_names_out())
cv_df.head()
```

```python
# Train test split
X_cv = cv_df
y_cv = data['Label']
X_train_cv, X_test_cv, y_train_cv, y_test_cv = train_test_split(X_cv, y_cv, test_size=0.2, random_state=36)
```

```python
# Support Vector Machine
cv_svm = SVC()

cv_svm.fit(X_train_cv, y_train_cv)

cv_svm_predict = cv_svm.predict(X_test_cv)

cv_svm_accuracy = accuracy_score(y_test_cv, cv_svm_predict)

print(cv_svm_accuracy)
print(classification_report(y_test_cv, cv_svm_predict))
```

The classification report above shows that SVM gives an overall pretty good accuracy (71%). But we also can see that the 'Politics' recall metrics is so low (33%) and it tells that the model can classify 'Politics' well but still missing a lot of detecting the 'Politics' category.

```python
# Random Forest
cv_rf = RandomForestClassifier()

cv_rf.fit(X_train_cv, y_train_cv)

cv_rf_predict = cv_rf.predict(X_test_cv)

cv_rf_accuracy = accuracy_score(y_test_cv, cv_rf_predict)

print(cv_rf_accuracy)
print(classification_report(y_test_cv, cv_rf_predict))
```

The classification report above shows that Random Forest gives an overall high accuracy (86%). The classification report metrics also shows a high number and it indicated that the Random Forest model performs pretty well. The lowest number is on the 'Olahraga' recall metrics and it indicates that the model is still missing a lot of detecting the 'Sports' category.
